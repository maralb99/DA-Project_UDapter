```python
# Function to load JSON files
def load_json(file_path):
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File not found: {file_path}")
    with open(file_path, 'r') as file:
        return json.load(file)
```

```python
# Metric to compare
metric = 'precision'
categories = ["Tokens", "Sentences", "Words", "UPOS", "XPOS", "UFeats", "AllTags", "Lemmas", "UAS", "LAS", "CLAS", "MLAS", "BLEX"]

# Function to extract the relevant data from the JSON dictionary
def extract_metric(data, metric, categories):
    extracted_data = []
    for category in categories:
        extracted_data.append(data[category].get(metric, 0))
    return extracted_data
```

## Significance test (z-test)
The graph shows that FTA_80 and ITA_80 perform similarly well on the Georgian test set. 
To see whether the differences are statistically significant, the z-test for the precision score 
in the categories UAS, LAS and CLAS is performed here. 

```python
import numpy as np
from scipy.stats import norm
```

```python
#function for z-test
def precision_difference_test(precision1, n1, precision2, n2):
    SE = np.sqrt(precision1 * (1 - precision1) / n1 + precision2 * (1 - precision2) / n2)
    z_score = (precision1 - precision2) / SE
    p_value = 2 * (1 - norm.cdf(np.abs(z_score)))
    
    return p_value
```

```python
#P_value for UAS
precision_fta_80_G_UAS = 0.7400428265524626
precision_ita_80_G_UAS = 0.7310492505353319
n_sentences_G_test_set_f = 2335
n_sentences_G_test_set_i = 2335

p_value = precision_difference_test(precision_fta_80_G_UAS, n_sentences_G_test_set_f, precision_ita_80_G_UAS, n_sentences_G_test_set_i)
print(f"P(UAS)= {p_value}")
```

```python
#P_value for LAS
precision_fta_80_G_LAS = 0.5691648822269807
precision_ita_80_G_LAS = 0.5618843683083512
n_sentences_G_test_set_f = 2335
n_sentences_G_test_set_i = 2335

p_value = precision_difference_test(precision_fta_80_G_LAS, n_sentences_G_test_set_f, precision_ita_80_G_LAS, n_sentences_G_test_set_i)
print(f"P(LAS)= {p_value}")
```

```python
#P_value for CLAS = 0.5691648822269807
precision_fta_80_G_CLAS = 0.5497887748943875
precision_ita_80_G_CLAS = 0.5335797905113987
n_sentences_G_test_set_f = 2335
n_sentences_G_test_set_i = 2335

p_value = precision_difference_test(precision_fta_80_G_CLAS, n_sentences_G_test_set_f, precision_ita_80_G_CLAS, n_sentences_G_test_set_i)
print(f"P(CLAS)= {p_value}")
```

Since the P-value must be less than 0.05 (P>0.05) to be considered statistically relevant, 
it can be said that none of the differences in the precision score are statistically significant. 
This means that the FTA model does not perform significantly better on the Georgian test set than 
the ITA model. This means that the hypothesis that the FTA model performs better on the Georgian 
test set than the ITA model can be refuted.

```python
#Looking for the search for the five simplest and five most complex sentences that have been annotated most differently
def load_conllu(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        sentences = []
        current_sentence = []
        for line in file:
            line = line.strip()
            if line.startswith('#'):
                continue
            if line == '':
                if current_sentence:
                    sentences.append(current_sentence)
                    current_sentence = []
                continue
            token_data = line.split('\t')
            current_sentence.append(token_data)
        if current_sentence:
            sentences.append(current_sentence)
    return sentences

def compare_conllu(FTA_80, ITA_80):
    sentences1 = load_conllu(FTA_80)
    sentences2 = load_conllu(ITA_80)
    
    differences = []
    max_len = max(len(sentences1), len(sentences2))
    for i in range(max_len):
        if i >= len(sentences1):
            differences.append((i+1, None, sentences2[i]))
            continue
        if i >= len(sentences2):
            differences.append((i+1, sentences1[i], None))
            continue
        
        sent1 = sentences1[i]
        sent2 = sentences2[i]
        if len(sent1) != len(sent2):
            differences.append((i+1, sent1, sent2))
            continue
        
        for j, (token1, token2) in enumerate(zip(sent1, sent2)):
            if token1 != token2:
                differences.append((i+1, sent1, sent2))
                break
    
    return differences

def sentence_complexity(sentence):
    return len(sentence)

def extract_sentence_text(sentence):
    sentence_sorted = sorted(sentence, key=lambda x: float(x[0].split('-')[0]))
    return ' '.join([token[1] for token in sentence_sorted if '-' not in token[0]])

# File paths
FTA_80 = 'FTA_80_Georgian_output.conllu'
ITA_80 = 'ITA_80_Georgian_output.conllu'

# Compare the files
differences = compare_conllu(FTA_80, ITA_80)

# Sort differences by sentence complexity
differences_sorted = sorted(differences, key=lambda x: sentence_complexity(x[1]) if x[1] is not None else 0)

# Get the 5 simplest and 5 most complex differing sentences
simplest_differences = differences_sorted[:5]
most_complex_differences = differences_sorted[-5:]

# Print the results
print("Simplest differing sentences:")
for diff in simplest_differences:
    i, sent1, sent2 = diff
    print(f"Sentence {i}:")
    print(f"FTA_80: {extract_sentence_text(sent1) if sent1 else 'MISSING'}")
    print(f"ITA_80: {extract_sentence_text(sent2) if sent2 else 'MISSING'}")
    print()

print("Most complex differing sentences:")
for diff in most_complex_differences:
    i, sent1, sent2 = diff
    print(f"Sentence {i}:")
    print(f"FTA_80: {extract_sentence_text(sent1) if sent1 else 'MISSING'}")
    print(f"ITA_80: {extract_sentence_text(sent2) if sent2 else 'MISSING'}")
    print()
```

```python
import spacy
from spacy.tokens import Doc
from spacy import displacy
from IPython.display import display, HTML
import re

def visualize_dependency_tree_with_title(sentence, title, language='xx'):
    # Load the spaCy model for the given language
    nlp = spacy.blank(language)
    
    # Split the sentence into lines and tokens
    lines = sentence.strip().split('\n')
    words = []
    heads = []
    deps = []
    
    for line in lines:
        parts = line.split('\t')
        words.append(parts[1])
        heads.append(int(parts[6]) - 1)  # 0-based index for head
        deps.append(parts[7])
    
    # Create the spaCy Doc and tokens
    doc = Doc(nlp.vocab, words=words)
    for word, head, dep in zip(doc, heads, deps):
        word.dep_ = dep
        if head >= 0:  # handle root case where head is '0'
            word.head = doc[head]
    
    # Render the Dependency Parser Tree
    svg = displacy.render(doc, style='dep', options={'compact': True}, jupyter=False, page=True)
    
    # Find the root token index
    root_token = [token for token in doc if token.dep_ == 'root'][0]
    root_text = root_token.text
    
    # Modify the SVG to highlight the root node
    # The SVG has <text class="word"> elements, we'll search for the one with the root word
    highlighted_svg = re.sub(
        fr'(<text class="word" fill="currentColor" text-anchor="middle" y=".*?" data-ord=".*?">{root_text}</text>)',
        r'<text class="word" fill="red" font-weight="bold" text-anchor="middle" y=".*?" data-ord=".*?">{root_text}</text>',
        svg
    )
    
    # Create HTML with title above the tree and modified SVG
    html_with_title = f'''
    <h3 style="text-align: center;">{title}</h3>
    {highlighted_svg}
    '''
    
    return html_with_title
```

```python
#function for complex dependency trees
def visualize_dependency_tree_with_title(sentence, title, language='xx'):
    # Load the spaCy model for the given language (in this case, Georgian)
    nlp = spacy.blank(language)
    
    # Split the sentence into lines and tokens
    lines = sentence.strip().split('\n')
    words = []
    heads = []
    deps = []
    
    for line in lines:
        parts = line.strip().split('\t')
        if len(parts) >= 8 and parts[0].isdigit():  # Check if it's a valid line
            words.append(parts[1])
            head = int(parts[6]) - 1 if parts[6].isdigit() else None  # 0-based index for head
            heads.append(head)
            deps.append(parts[7])
    
    # Create the spaCy Doc and tokens
    doc = Doc(nlp.vocab, words=words)
    for token_idx, (head, dep) in enumerate(zip(heads, deps)):
        if head is not None and head >= 0:
            doc[token_idx].head = doc[head]
            doc[token_idx].dep_ = dep
    
    # Render the Dependency Parser Tree
    html = displacy.render(doc, style='dep', options={'compact': True}, jupyter=False)
    
    # Create HTML with title above the tree
    html_with_title = f'''
    <h3 style="text-align: center;">{title}</h3>
    {html}
    '''
    
    return html_with_title
```

- both models give similar results

Most common grammatical issues:

Subject and Object: Instances such as "დავითი" (David) being marked as an object (obj) instead of a nominal subject (nsubj), and "ურთიერთობა" (relationship) labeled as obj instead of nsubj.

Nominal Modifiers, Possessives and Determiner: Confusion arises between nmod (nominal modifier), nmod:poss

```python
# Initialize variables to store total mistakes and total sentences
total_fta_mistakes = 0
total_ita_mistakes = 0
total_sentences = 0

# Iterate through each sentence
for i in range(len(data["GLC_ID"])):
    tokens = data["Quantity of tokens"][i]
    fta_mistakes = data["FTA_80"][i]
    ita_mistakes = data["ITA_80"][i]
    
    if tokens is not None and fta_mistakes is not None:
        avg_fta_per_sentence = fta_mistakes / tokens
        total_fta_mistakes += avg_fta_per_sentence
    
    if tokens is not None and ita_mistakes is not None:
        avg_ita_per_sentence = ita_mistakes / tokens
        total_ita_mistakes += avg_ita_per_sentence
    
    if tokens is not None:  # Count valid sentences
        total_sentences += 1

# Calculate overall averages
average_fta_mistakes_per_sentence = total_fta_mistakes / total_sentences
average_ita_mistakes_per_sentence = total_ita_mistakes / total_sentences

# Print results
print(f"Average mistakes per sentence for FTA_80: {average_fta_mistakes_per_sentence:.3f}")
print(f"Average mistakes per sentence for ITA_80: {average_ita_mistakes_per_sentence:.3f}")
```
